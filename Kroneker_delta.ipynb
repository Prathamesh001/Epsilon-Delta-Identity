```python
def epsilon(i, j, k):
    # If any two indices are the same, the value is 0.
    if i == j or j == k or k == i:
        return 0

    # Positive permutations (even permutations)
    elif (i, j, k) == (1, 2, 3) or (i, j, k) == (2, 3, 1) or (i, j, k) == (3, 1, 2):
        return 1

    # Negative permutations (odd permutations)
    elif (i, j, k) == (3, 2, 1) or (i, j, k) == (1, 3, 2) or (i, j, k) == (2, 1, 3):
        return -1

    # This 'else' case is technically redundant due to the first 'if' statement
    else:
        return 0

def delta(a, b):
  if a == b:
    return 1
  else:
    return 0
sum = 0
print("Non Zero Terms:")
for j in range(1,4):
  for k in range(1,4):
    for m in range(1,4):
      for n in range(1,4):
        for i in range(1,4):
          #if epsilon(i,j,k)*epsilon(i,m,n) !=0:
          print(f'ε{i,j,k}ε{i,m,n} = {epsilon(i,j,k)} x {epsilon(i, m, n)} = {epsilon(i,j,k)*epsilon(i,m,n)}')
          sum+= epsilon(i,j,k)*epsilon(i,m,n)
        #if sum != 0:
        print(f'Hence, LHS => ε1{j}{k}ε1{m}{n} + ε2{j}{k}ε2{m}{n} + ε3{j}{k}ε3{m}{n} = {sum}')
        print(f'         RHS => δ{j, m}δ{k,n} - δ{j, n}δ{k, m} = {delta(j, m)} x {delta(k, n)} - {delta(j, n)} x {delta(k, m)} = {delta(j, m)*delta(k, n) - delta(j, n)*delta(k, m)}')
        sum = 0
```
